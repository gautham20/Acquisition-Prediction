```python
import pandas as pd
import numpy as np
from datetime import *
```

```python
companies = pd.read_csv('comp_new.csv')
```

```python
companies = list(companies['name'])
```

```python
features = ['acquired','founded','nemps','nactivities','nfollowers','noffices','nfundrounds'
           ,'fundamt','ninvestors','fundamts','hq','revenue','ceo_app','owl_nfollow','pro_comp'
            ,'twiiter','fb']
len(features)
```

```python
feature_dict = {}
ncomp = 0
```

```python
len(feature_dict)
```

```python
def get_compdict(comp):
    global features
    compdict = {}
    compdict['name'] = comp
    print(comp)
    for f in features:
        compdict[f] = raw_input(f+':')
    return compdict

for comp in companies[ncomp:min(ncomp+6,len(companies))]:
    feature_dict[comp] = get_compdict(comp)

ncomp+=6
```

```python
#feature_dict
f_df = pd.DataFrame(feature_dict)
f_df = f_df.T
f_df.to_pickle('features_new.pkl')
```
